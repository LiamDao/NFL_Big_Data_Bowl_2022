```python
# Import packages, set working directory, and set views
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
import os
import matplotlib.pyplot as plt
import seaborn as sns
os.chdir('C:\\Users\\cmcle\\Documents\\NFL Big Data Bowl\\')
```

```python
# Import data
punts2018 = pd.read_csv('punts2018.csv')
punts2019 = pd.read_csv('punts2019.csv')
punts2020 = pd.read_csv('punts2020.csv')
```

```python
# Subset on Panthers
car2018 = punts2018[punts2018['possessionTeam'] == 'CAR']
car2019 = punts2019[punts2019['possessionTeam'] == 'CAR']
car2020 = punts2020[punts2020['possessionTeam'] == 'CAR']
car2020.head()
```

```python
# Sum missing values across all columns for 2018
car2018.isnull().sum().sort_values(ascending = False)
```

```python
# Sum missing values across all columns for 2019
car2019.isnull().sum().sort_values(ascending = False)
```

```python
# Sum missing values across all columns for 2020
car2020.isnull().sum().sort_values(ascending = False)
```

```python
# Drop duplicates of game ID and play ID otherwise duplicate play observations
car2018_nodup = car2018.drop_duplicates(['gameId', 'playId'])
car2019_nodup = car2019.drop_duplicates(['gameId', 'playId'])
car2020_nodup = car2020.drop_duplicates(['gameId', 'playId'])
car2020_nodup.head()
```

```python
# Average return yardage by quarter for 2018
car2018_nodup.groupby('quarter').agg({'kickReturnYardage' : 'mean'}).plot.bar(rot = 0)
```

```python
# Average return yardage by quarter for 2019
car2019_nodup.groupby('quarter').agg({'kickReturnYardage' : 'mean'}).plot.bar(rot = 0)
```

```python
# Average return yardage by quarter for 2020
car2020_nodup.groupby('quarter').agg({'kickReturnYardage' : 'mean'}).plot.bar(rot = 0)
```

```python
# Average play result by quarter for 2018
    # Yards gained by kicking team
car2018_nodup.groupby('quarter').agg({'playResult' : 'mean'})
```

```python
# Average play result by quarter for 2019
    # Yards gained by kicking team
car2019_nodup.groupby('quarter').agg({'playResult' : 'mean'})
```

```python
# Average play result by quarter for 2020
    # Yards gained by kicking team
car2020_nodup.groupby('quarter').agg({'playResult' : 'mean'})
```

```python
# Average hang time by quarter for 2018
car2018_nodup.groupby('quarter').agg({'hangTime' : 'mean'})
```

```python
# Average hang time by quarter for 2019
car2019_nodup.groupby('quarter').agg({'hangTime' : 'mean'})
```

```python
# Average hang time by quarter for 2020
    # Is this all punt data?
car2020_nodup.groupby('quarter').agg({'hangTime' : 'mean'})
# car2020_nodup.groupby('quarter').size()
```

```python
# Correlation plot for 2018
    # High correlation between kick return yardage and penalty yards
plt.figure(figsize = (19, 10))
sns.heatmap(car2018_nodup.drop(['frameId', 'kickoffReturnFormation', 'season', 'kickerId', 
                                'kickBlockerId', 'down', 'jerseyNumber', 'gameId',
                               'playId', 'Unnamed: 0'], axis = 1).corr())
plt.title('Correlation Heatmap')
plt.show()
```
